```python
data = [("iPhone", 999), ("Samsung", 799), ("MacBook", 1299)]
df = spark.createDataFrame(data, ["product", "price"])
df.show()
```

```python
df.filter(df.price > 1000).show()
```

```python
df.printSchema()
```

```python
class_data = [("Arjun", 999,"DS"), ("Siva", 799,"BS"), ("Veer", 1299,"BS"), ("Raj",459,"CSE")]
df = spark.createDataFrame(class_data, ["name", "id", "branch"])
df.show()
```

```python
df.filter(df.branch == "BS").show()
```
