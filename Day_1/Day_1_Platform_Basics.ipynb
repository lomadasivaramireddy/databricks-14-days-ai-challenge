**🌍Real-World Example – DAY 01 (🏢Industry-Style)**

🚦 Smart City – Traffic Congestion _Analysis_

```python
data = [
    ("S101", "Main Road", "Morning", 1200, 35),
    ("S102", "Ring Road", "Morning", 3400, 18),
    ("S103", "Highway", "Morning", 5200, 70),
    ("S104", "Main Road", "Evening", 2900, 22),
    ("S105", "Ring Road", "Evening", 4100, 15)
]

columns = [
    "sensor_id",
    "road_name",
    "time_slot",
    "vehicle_count",
    "avg_speed_kmph"
]

traffic_df = spark.createDataFrame(data, columns)
traffic_df.show(truncate=False)
```

```python
congested_df = traffic_df.filter(
    (traffic_df.vehicle_count > 3000) & 
    (traffic_df.avg_speed_kmph < 25)
)

congested_df.show(truncate=False)
```

```python
traffic_df.createOrReplaceTempView("traffic_data")
```

**SQL Queries**
**🚦 Smart City – Traffic Congestion Analysis (Industry Use Case)**

**🚨 1. High-Congestion Roads**

```python
spark.sql("""
SELECT road_name, time_slot, vehicle_count, avg_speed_kmph
FROM traffic_data
WHERE vehicle_count > 3000 AND avg_speed_kmph < 25
""").show(truncate=False)
```

**📊 2. Average Speed Per Road**

```python
spark.sql("""
SELECT road_name, ROUND(AVG(avg_speed_kmph), 2) AS avg_speed
FROM traffic_data
GROUP BY road_name
""").show()
```

**📈 3. Traffic Volume by Time Slot**

```python
spark.sql("""
SELECT time_slot, SUM(vehicle_count) AS total_vehicles
FROM traffic_data
GROUP BY time_slot
""").show()
```
