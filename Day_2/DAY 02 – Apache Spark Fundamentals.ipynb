```python
events = spark.read.csv(
    "/Volumes/workspace/ecommerce/ecommerce_data/*.csv",
    header=True,
    inferSchema=True
)

events.limit(10).show()
```

```python
events.printSchema()
```

**1️⃣ select**

```python
events.select("event_type", "brand", "price").limit(10).show()
```

**2️⃣ filter**

```python
events.filter("price > 100").count()
```

**3️⃣ groupBy**

```python
events.groupBy("event_type").count().show()
```

**4️⃣ orderBy (Top brands)**

```python
top_brands = (
    events
    .groupBy("brand")
    .count()
    .orderBy("count", ascending=False)
)

top_brands.show(10)
```

**Export Results (WRITE BACK TO VOLUME)**

```python
top_brands.write.mode("overwrite").csv(
    "/Volumes/workspace/ecommerce/ecommerce_data/day02_top_brands"
)
```
